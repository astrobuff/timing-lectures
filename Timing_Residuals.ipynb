````python
T.data = os.getcwd() # Directory where the parameter file is found
# a bunch of observations, randomly taken
obstimes = N.random.uniform(56600,57200,100)
obstimes.sort()

# Fake the arrival times!
psr = LT.fakepulsar(parfile=os.path.join(T.data,'pulsar.par'),
                    obstimes=obstimes,
                    toaerr=100)

# Model observing times and pulsar
model_obstimes = N.linspace(56600,57200,6001)
model_psr = LT.fakepulsar(parfile=os.path.join(T.data,'pulsar.par'),
                          obstimes=model_obstimes,
                          toaerr=0.0001)

LT.make_ideal(model_psr)

LT.add_efac(psr,efac=1.0)

# save these values 
psr.savetim('./pulsar.tim')
model_psr.savetim('./model.tim')

# Convenience function to plot data and model
def plot_models_and_errors(psr, model_psr, xrange=None, ax=None, title=None):
    res, t, errs = psr.residuals(), psr.toas(), psr.toaerrs
    mres, mt, merrs = model_psr.residuals(), model_psr.toas(), model_psr.toaerrs

    if ax is None:
        ax = plt.gca()
    ax.plot(mt, mres, lw=2, color='r', zorder=1)
    ax.errorbar(t, res, yerr=errs / 1e6, color='b', fmt='none', zorder=10)
    
    if xrange is not None:
        ax.set_xlim(xrange)
    if title is not None:
        ax.set_title(title)

plot_models_and_errors(psr, model_psr)
````

## Let's play!
Now we're going to modify slightly single parameters of the timing model, and we'll see the kind of residuals that will come out.
First of all, we start with a very slightly different value of the pulse frequency (one part in ten billions, or ~$7\times10^{-9}$ Hz). 
Remember that 
\begin{equation}
\phi(t) = \phi_0 + \nu t + \frac{1}{2} \dot{\nu} t^2 + \frac{1}{6} \ddot{\nu} t^3 + \dots
\end{equation}
So, if instead of $\nu$ we use $\tilde{\nu}$, we expect a new $\tilde{\phi}$ that differs from $\phi$ of a quantity $\Delta\phi = \tilde{\phi} - \phi = (\tilde{\nu} - \nu) t$. This means that the residuals will have an additional _linear_ trend.

````python
psr = T.tempopulsar(parfile = os.path.join(T.data,'pulsar.par'),
                    timfile = os.path.join(T.data,'pulsar.tim'))
model_psr = T.tempopulsar(parfile = os.path.join(T.data,'pulsar.par'),
                          timfile = os.path.join(T.data,'model.tim'))


````

````python
rel_error = 1e-10
new_F0 = psr['F0'].val * (1 - rel_error)
psr['F0'].val = new_F0
model_psr['F0'].val = new_F0
print('dF = {} Hz'.format(rel_error * psr['F0'].val))

plot_models_and_errors(psr, model_psr)
````

## Effect of frequency derivative

Now let us try with an error on the frequency derivative. We use a mismatch of ~$5\cdot 10^{-16}$ Hz/s. According to the points above, this wrong frequency derivative will produce a spurious term proportional to $(\tilde{\dot{\nu}} - \dot{\nu}) t^2$, so a _quadratical_ term.

````python
psr = T.tempopulsar(parfile = os.path.join(T.data,'pulsar.par'),
                    timfile = os.path.join(T.data,'pulsar.tim'))
model_psr = T.tempopulsar(parfile = os.path.join(T.data,'pulsar.par'),
                          timfile = os.path.join(T.data,'model.tim'))

dFdot = -5e-16
newval = psr['F1'].val + dFdot
psr['F1'].val = newval
model_psr['F1'].val = newval


plot_models_and_errors(psr, model_psr)
````

## Effect of a wrong orbital period
A slightly wrong orbital period will produce structure in the residuals proportional to

\begin{align}
\tilde{\phi} - \phi &= A [\sin(\tilde{\Omega} t) - \sin(\Omega t)] \\
                    &= 2A [\sin(\alpha t)\cos(\beta t)]
\end{align}

where $A$ is a constant, $\Omega = 2\pi / P_{orb}$ is the orbital angular velocity, $\alpha = (\tilde{\Omega} - \Omega) / 2$ and $\beta = (\tilde{\Omega} + \Omega) / 2 \approx \Omega$. So, it will be a sinusoid at a very low frequency (the difference between the right and the wrong orbital frequencies), modulated by another sinusoid at approximately the orbital frequency. The smaller the error, the lower the frequency of this modulation.

````python
psr = T.tempopulsar(parfile = os.path.join(T.data,'pulsar.par'),
                    timfile = os.path.join(T.data,'pulsar.tim'))

model_psr = T.tempopulsar(parfile = os.path.join(T.data,'pulsar.par'),
                          timfile = os.path.join(T.data,'model.tim'))

from matplotlib import gridspec
fig = plt.figure(figsize=(12,6))
dPB = 2e-7

fig.suptitle('P$_{orb}$ uncertainty = ' + '{} ms'.format(dPB * 86400 * 1000))
gs = gridspec.GridSpec(1,3)
ax = plt.subplot(gs[0])
print('dPB = {} ms'.format(dPB * 86400 * 1000))
newval = psr['PB'].val - dPB
psr['PB'].val = newval
model_psr['PB'].val = newval

plot_models_and_errors(psr, model_psr, ax=ax, title='Full')

ax = plt.subplot(gs[1])
plot_models_and_errors(psr, model_psr, ax=ax, xrange=[56700, 56750], title='Early Zoom')

ax = plt.subplot(gs[2])
plot_models_and_errors(psr, model_psr, ax=ax, xrange=[57100, 57130], title='Late Zoom')
````

## Effect of a wrong $a \sin i$
In this case, as we expect, we'll have a spurious sinusoidal modulation due to the non complete correction, or the over correction, of the orbital modulation. The amplitude of this modulation will be constant.

````python
psr = T.tempopulsar(parfile = os.path.join(T.data,'pulsar.par'),
                    timfile = os.path.join(T.data,'pulsar.tim'))

model_psr = T.tempopulsar(parfile = os.path.join(T.data,'pulsar.par'),
                          timfile = os.path.join(T.data,'model.tim'))
newval = psr['A1'].val * 0.99995
psr['A1'].val = newval
model_psr['A1'].val = newval


res, t, errs = psr.residuals(), psr.toas(), psr.toaerrs

fig = plt.figure(figsize=(8,6))
gs = gridspec.GridSpec(1,2)
ax = plt.subplot(gs[0])
plot_models_and_errors(psr, model_psr, ax=ax, xrange=[56700, 56750], title='Early Zoom')
ax = plt.subplot(gs[1])
plot_models_and_errors(psr, model_psr, ax=ax, xrange=[57100, 57130], title='Late Zoom')
````

## Wrong $T_0$
Let's add one _second_ to $T_0$. 
For a small addition $\delta$ to $T_0$, we have something like
\begin{align}
\tilde{\phi} - \phi &= A [\sin(\Omega(t + \delta)) - \sin(\Omega t)] \\
                    &= A [\sin(\Omega t)\cos(\Omega \delta) + \cos(\Omega t)\sin(\Omega \delta) - \sin(\Omega t)]\\
                    &\approx A \cos(\Omega t) \epsilon
\end{align}
where $\epsilon=\sin(\Omega \delta)$ is a constant quantity that drives the amplitude of the spurious modulation

<p></p>

<p></p>

<div style='font-size: 80%; .MathJax {font-size: 80%;}'>The sum $\sin(\Omega t)\cos(\Omega \delta) - \sin(\Omega t)$ vanishes approximately like $\sin^2(\Omega \delta)$</div>

````python
psr = T.tempopulsar(parfile = os.path.join(T.data,'pulsar.par'),
                    timfile = os.path.join(T.data,'pulsar.tim'))

model_psr = T.tempopulsar(parfile = os.path.join(T.data,'pulsar.par'),
                          timfile = os.path.join(T.data,'model.tim'))
#I add one second to T0!
newval = psr['T0'].val + 1 / 3600 / 24
psr['T0'].val = newval
model_psr['T0'].val = newval

res, t, errs = psr.residuals(), psr.toas(), psr.toaerrs

fig = plt.figure(figsize=(8,6))
gs = gridspec.GridSpec(1,2)
ax = plt.subplot(gs[0])
plot_models_and_errors(psr, model_psr, ax=ax, xrange=[56700, 56750], title='Early Zoom')
ax = plt.subplot(gs[1])
plot_models_and_errors(psr, model_psr, ax=ax, xrange=[57100, 57150], title='Late Zoom')
````
